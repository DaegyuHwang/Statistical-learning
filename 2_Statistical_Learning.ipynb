```python
college = pd.read_csv('College.csv')
college
```

```python
college = college.set_index('Unnamed: 0') # deles the column and set as an index # .set_index('')
college.index.name = 'Names' # change index name
college.head() # show first few rows
```

```python
college.describe(include='all')
```

```python
college.shape
```

```python
g = sns.PairGrid(college, vars=college.iloc[:,1:11], hue='Private')
g.map_upper(plt.scatter,s=3)
g.map_diag(plt.hist)
g.map_lower(plt.scatter,s=3)
g.fig.set_size_inches(12,12)
```

```python
sns.boxplot(x='Private', y='Outstate', data=college); # boxplot
```

```python
college.loc[college['Top10perc']>50,'Elite']='Yes' # making new column by binning one column
college['Elite']=college['Elite'].fillna('No')

sns.boxplot(x='Elite',y='Outstate',data=college);
```

```python
# Bin creation
college['PhD'] = pd.cut(college['PhD'],3,labels=['Low','Medium','High'])
college['Grad.Rate'] = pd.cut(college['Grad.Rate'],5,labels=['Very low', 'Low', 'Medium', 'High', 'Very high'])
college['Books'] = pd.cut(college['Books'], 2, labels=['Low', 'High'])
college['Enroll'] = pd.cut(college['Enroll'], 4, labels=['Very low', 'Low', 'High', 'Very high'])
```

```python
fig = plt.figure()

plt.subplot(221) # 2rows, 2columns, and position 1. # matlab ìƒê°
college['PhD'].value_counts().plot(kind='bar',title='Private');
plt.subplot(222)
college['Grad.Rate'].value_counts().plot(kind='bar',title='Grad.Rate')
plt.subplot(223)
college['Books'].value_counts().plot(kind='bar', title = 'Books');
plt.subplot(224)
college['Enroll'].value_counts().plot(kind='bar', title = 'Enroll');

fig.subplots_adjust(hspace=1) # to add space between subplots
```

```python
data_url = 'https://raw.githubusercontent.com/scikit-learn/scikit-learn/main/sklearn/datasets/data/boston_house_prices.csv'
df = pd.read_csv(data_url, sep=',',header=1) # Set the second row from the data as a name of column
df
```

```python
df.shape # 506 rows 14 columns
```

```python
df = df.rename({'MEDV':'target'},axis=1)
df.head()
```

```python
g = sns.PairGrid(df)
g.map_upper(plt.scatter,s=3)
g.map_diag(plt.hist)
g.map_lower(sns.kdeplot, cmap='Blues_d')
g.fig.set_size_inches(12,12)
```

```python
g = sns.PairGrid(df)
g.map_upper(plt.scatter,s=3)
g.map_diag(plt.hist)
g.map_lower(plt.scatter,s=3)
g.fig.set_size_inches(12,12)
```

```python
# focusing on one plot # exist a positive linear relationship
plt.scatter(df['RM'], df['target'])
plt.xlabel('RM')
plt.ylabel('target');
```

```python
# It seems to have a negative non-linear relationship
plt.scatter(df['LSTAT'], df['target'])
plt.xlabel('LSTAT')
plt.ylabel('target');
```

```python
# It seems to exist a negative non-linear relationship
plt.scatter(df['RM'], df['LSTAT'])
plt.xlabel('RM')
plt.ylabel('LSTAT');
```

```python
# investigate the correlation of each variable # showing closer relationship with RAD,TAX,LSTAT
df.corrwith(df['CRIM']).sort_values()
```

```python
# we found that twhen RAD is equal to 24, average CRIM is much higher and CRIM range is much larger
ax = sns.boxplot(x='RAD',y='CRIM',data=df)
```

```python
# we found that when TAX is equal to 666, average CRIM is much higher and CRIM range is much larger
plt.scatter(df['TAX'],df['CRIM'])
plt.xlabel('TAX')
plt.ylabel('CRIM');
```

```python
# For lower values of LSTAT (< 10), CRIM is always under 10. For LSTAT > 10, there is a wider spread of CRIM.
# For LSTAT < 20, a large proportion of the data points is very close to CRIM = 0.
plt.scatter(df['LSTAT'], df['CRIM'])
plt.xlabel('LSTAT')
plt.ylabel('CRIM');
```
