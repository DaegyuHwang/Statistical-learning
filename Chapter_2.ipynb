```python
import numpy as np
import pandas as pd

college = pd.read_csv('College.csv')
college
```

```python
# index_col=0: set the first column as the idex of data
college2=pd.read_csv('College.csv',index_col=0)
college2
```

```python
# change the column name
college3=college.rename({'Unnamed: 0': 'College'}, axis=1)
college3
```

```python
# set a column as a index
college3=college3.set_index('College')
college3
```

```python
college = college3
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])
```

```python
from matplotlib.pyplot import subplots

fig, ax = subplots(figsize=(8,8))
college.boxplot('Outstate', by='Private', ax=ax)
```

```python
# set new column by cutting one column depends on their percentage
college['Elite']=pd.cut(college['Top10perc'],[0,50,100],labels=['No','Yes'])
college['Elite'].value_counts()
```

```python
fix, ax = subplots(figsize=(8,8))
college.boxplot('Outstate', by='Elite', ax=ax)
```

```python
fig, ax = subplots(figsize=(8,8))
college.hist('Top10perc', ax=ax)
```

```python
g = sns.PairGrid(college, vars=college.iloc[:,1:11], hue='Private')
g.map_upper(plt.scatter, s=3)
g.map_diag(plt.hist)
g.map_lower(plt.scatter, s=3)
g.fig.set_size_inches(12, 12)
```

```python
# deleting missing data
Auto=pd.read_csv('Auto.csv', na_values=['?'])
Auto=Auto.dropna()
Auto=Auto.set_index('name')
```

```python
## Quantitative : types like 'int64' or 'float64'
## Qualitative: types like 'object' or 'category'
Auto.dtypes
```

```python
## min max mean std
print(Auto['mpg'].min(), Auto['mpg'].max())
print(Auto['year'].min(), Auto['year'].max())
print(Auto['year'].mean(), Auto['year'].std())
```

```python
Auto_subset = Auto.drop(Auto.index[9:85])
print(Auto_subset['year'].min(), Auto_subset['year'].max())
print(Auto_subset['year'].mean(), Auto_subset['year'].std())
```
